```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a dataset with a normal distribution
np.random.seed(0)
data = np.random.normal(50, 10, 1000)

# Introduce outliers into the dataset
data_with_outliers = np.append(data, [500, 600, 700])

# Calculate mean and median for both datasets
mean_original = np.mean(data)
median_original = np.median(data)
mean_with_outliers = np.mean(data_with_outliers)
median_with_outliers = np.median(data_with_outliers)

print("Original Data:")
print(f"Mean: {mean_original}")
print(f"Median: {median_original}\n")

print("Data with Outliers:")
print(f"Mean: {mean_with_outliers}")
print(f"Median: {median_with_outliers}\n")

# Plot the data to visualize the effect of outliers
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='blue', alpha=0.7, label='Original Data')
plt.axvline(mean_original, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_original:.2f}')
plt.axvline(median_original, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_original:.2f}')
plt.legend()
plt.title('Original Data')

plt.subplot(1, 2, 2)
plt.hist(data_with_outliers, bins=30, color='orange', alpha=0.7, label='Data with Outliers')
plt.axvline(mean_with_outliers, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_with_outliers:.2f}')
plt.axvline(median_with_outliers, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_with_outliers:.2f}')
plt.legend()
plt.title('Data with Outliers')

plt.tight_layout()
plt.show()
```

Z - SCORE

```python
import numpy as np
import pandas as pd
```

```python
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data) # Indentation corrected to 8 spaces
    return (data - mean) / std
```

```python
z_scores = calculate_z_scores(data)
```

```python
def calculate_z_scores_pandas(data):
  """
  Calculates z-scores for a Pandas Series.

  Args:
      data: A Pandas Series containing the data.

  Returns:
      A Pandas Series containing the z-scores.
  """
  return pd.Series(data).transform(lambda x: (x - x.mean()) / x.std())
```

```python
z_scores = calculate_z_scores_pandas(data)
```

```python
import numpy as np

# Generate some sample data
data = np.random.normal(loc=5, scale=2, size=100)

# Calculate z-scores
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

# Filter data points within +- 3 standard deviations
filtered_data = data[(z_scores > -3) & (z_scores < 3)]

# Print the data points with z-scores between +- 3
print("Data points within +- 3 standard deviations:")
print(filtered_data)
```
