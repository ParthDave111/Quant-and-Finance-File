## 23.For the 5 strikes that your group member computed in Q15 and Q16, check whether put-call parity holds (within sensible rounding). Briefly comment on the reasons why/why not this is the case.

```python

# Define the put-call parity function for European options
def put_call_parity(call_prices, put_prices, S_ini, K, r, T):
    parity_values = []
    for i in range(len(K)):
        theoretical_diff = S_ini - K[i] * np.exp(-r * T)
        actual_diff = call_prices[i] - put_prices[i]
        parity_values.append((theoretical_diff, actual_diff))
    return parity_values


# Calculate put-call parity values
parity_values = put_call_parity(eur_call_prices, eur_put_prices, S_ini, K, r, T)

# Print the results
for i, k in enumerate(K):
    print(f"Strike Price: {k}")
    print(f"Theoretical Difference (S - K*e^(-rT)): {parity_values[i][0]:.2f}")
    print(f"Actual Difference (C - P): {parity_values[i][1]:.2f}")
    print(f"Difference: {abs(parity_values[i][0] - parity_values[i][1]):.2f}\n")
```

Comment:
The differences between the theoretical and actual values are relatively small, which indicates that put-call parity approximately holds, but not exactly. This is expected due to the early exercise feature of American options, which is not accounted for in the European put-call parity formula. The discrepancies arise because American options can be exercised at any time before expiration, affecting their pricing dynamics compared to their European counterparts.
