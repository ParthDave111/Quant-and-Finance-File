2. Student T

Student t copulas are particularly useful in situations where you need to model dependencies between multiple random variables with the potential for capturing tail dependencies and heavy-tailed behavior.[2]

```python
pycop.student().plot_pdf([0.3, 2], "3d", Nsplit=30)
```

Student t copulas is helpful  when you need to model dependencies with significant tail dependence and heavy-tailed behavior.

3. Clayton

Clayton copulas are particularly useful in situations where you need to model dependencies that exhibit strong lower tail dependence, meaning that the copula is more sensitive to joint occurrences of low values in the marginal distributions[2]

```python
pycop.archimedean(family="clayton").plot_pdf([2], "3d", Nsplit=30)
```

Clayton copulas is helpful when you need to model strong lower tail dependence and joint occurrences of low values in the marginal distributions.

4. Gumbel

Gumbel copulas are particularly useful in situations where you need to model dependencies that exhibit strong upper tail dependence, meaning that the copula is more sensitive to joint occurrences of high values in the marginal distributions[2]

```python
pycop.archimedean(family="gumbel").plot_pdf([2], "3d", Nsplit=30)
```

Gumbel copulas is helpful  when  model strong upper tail dependence and joint occurrences of high values in the marginal distributions.
