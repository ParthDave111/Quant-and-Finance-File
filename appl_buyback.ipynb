```python
import yfinance as yf
```

```python
ticker = "AAPL"
```

```python
apple_data = yf.download(ticker, start="2018-05-01", end="2024-05-04")
```

Apple Announced First Buy Back on May 1, 2018

Buy Back -100 Billion $

Analyzing return from May 1 2018 to April 30,2019

```python
First_buy_back = yf.download(ticker, start="2018-05-01", end="2019-04-30")
```

```python
First_buy_back
```

```python
daily_return = First_buy_back['Close'].pct_change()
```

```python
daily_return
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(daily_return, label='Daily Return')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.title('Daily Return of Apple Stock from May 1, 2018 to April 30, 2019')
plt.legend()
plt.show()
```

Assume I bought the stock on May 1st at $41.60

```python
purchase_price = 41.60
```

```python
latest_price = First_buy_back['Close'].iloc[-1]
```

```python
latest_price
```

```python
total_return = ((latest_price - purchase_price ) / purchase_price) * 100
```

```python
total_return
```
