Financial Health Check of Client - DTI ratio

```python
income = int(income)
```

```python
debt = input("Enter your current household debt  :")
```

```python
debt = int(debt)
```

```python
def calculate_dti(total_monthly_debt, gross_monthly_income):


  if gross_monthly_income == 0:
      return "Error: Gross monthly income cannot be zero."

  #  DTI
  dti = (total_monthly_debt / gross_monthly_income) * 100

  return dti


total_debt = debt
gross_income = income

dti_ratio = calculate_dti(total_debt, gross_income)

if isinstance(dti_ratio, str):
  print(dti_ratio)  # Print error message if applicable
else:
  print(f"Your debt-to-income ratio (DTI) is: {dti_ratio:.2f}%")
```

```python
if dti_ratio > 45:
  print("Your loan application is rejected due to high DTI ratio.")
else:
  print(f"Your DTI is within the acceptable limit. Your DTI is: {dti_ratio:.2f}%")
```

```python
import random

appraised_value = house_value * (0.90 + (1.10 - 0.90) * random.random())
print(appraised_value)
```

LOAN TO VALUE RATIO

```python
def calculate_ltv(loan_amount, property_value):

  if property_value == 0:

    return "Cannot calculate LTV. Property value cannot be zero."
  else:
    ltv = (loan_amount / property_value) * 100
    return round(ltv, 2)


loan_amount = loan_value
property_value = appraised_value


ltv = calculate_ltv(loan_amount, property_value)
print(f"The loan-to-value ratio (LTDR) is: {ltv:.2f}%")
```

```python
if ltv < 80:
  print("Congratulations! You are pre-approved for the mortgage.")
else:
  print("Sorry, we cannot provide you with the mortgage at this time.")
```
